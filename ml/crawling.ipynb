```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
```

```python
products1 = {
    '제품명' : [],
    '관리번호' : [],
    '성상' : [],
    '업체명' : [],
    '섭취방법' : [],
    '보관방법' : [],
    '주의사항' : [],
    '효능' : [],
    '기타' : []
}
```

```python
try:
    for idx in np.arange(3667, 4030):
        url1 = f'https://www.fmis.kr/public/pro/hpt/GbinHtfsPrductTrgetList.do?guPrdlstMnfManageNo=0&pageIndex={idx}&pageNo=4030600&searchCnd=99&searchWrd='
        df1 = pd.read_html(url1, encoding='utf-8')
        product_nums1 = df1[0]['품목 제조 관리번호']

        for product_num1 in product_nums1:
            info_url1 = f'https://www.fmis.kr/public/pro/hpt/selectHtfsPrductTrgetInfo.do?guPrdlstMnfManageNo={product_num1}&pageIndex={idx}&pageNo=4030600&searchCnd=99&searchWrd='
            headers = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',
                      'cookie' : 'JSESSIONID=3D0B66A897BF73D85A59B2EDD6C3960F.worker1; _gid=GA1.2.556744064.1686460481; _ga_H9T4K79JBD=GS1.1.1686482896.4.1.1686483468.0.0.0; _ga=GA1.2.1465316723.1686460481; _gat_gtag_UA_153717896_1=1',
                      'host' : 'www.fmis.kr',
                       'Accept-Language' : 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7'
                      }
            res1 = requests.get(info_url1, headers=headers, verify=False)
            soup1 = BeautifulSoup(res1.text, 'html.parser')
            try:
                board1 = soup1.select_one('div.board_view')
                products1['제품명'].append(board1.select_one('div.board_view.big > div > dl:nth-child(1) > dd').text)
                products1['관리번호'].append(board1.select_one('div.board_view.big > div > dl:nth-child(10) > dd').text)
                products1['성상'].append(board1.select_one('div.board_view.big > div > dl:nth-child(3) > dd').text)
                products1['업체명'].append(board1.select_one('div.board_view.big > div > dl:nth-child(4) > dd').text)
                products1['섭취방법'].append(board1.select_one('div.board_view.big > div > dl:nth-child(5) > dd').text)
                products1['보관방법'].append(board1.select_one('div.board_view.big > div > dl:nth-child(6) > dd').text)
                products1['주의사항'].append(board1.select_one('div.board_view.big > div > dl:nth-child(7) > dd').text)
                products1['효능'].append(board1.select_one('div.board_view.big > div > dl:nth-child(8) > dd').text)
                products1['기타'].append(board1.select_one('div.board_view.big > div > dl:nth-child(9) > dd').text)
            except:
                pass
            
except Exception as e:
    print(e)
    print(product_num1)
    print(idx)
```

```python
df = pd.DataFrame(products1)
df.drop_duplicates(inplace=True)
```

```python
df.to_csv('product.csv',encoding='utf-8', index=False, mode='a', header=False)
```
