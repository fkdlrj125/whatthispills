```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import pandas as pd
import os

options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
driver = webdriver.Chrome(options=options)

url = 'https://datalab.naver.com/shoppingInsight/sCategory.naver'
driver.get(url)
sleep(1)

# 버튼 설정
category1_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[7]/a'
category1 = driver.find_element(By.XPATH, category1_url)
driver.execute_script("arguments[0].click();", category1)
sleep(1)

category2_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[2]/ul/li[10]/a'
category2 = driver.find_element(By.XPATH, category2_url)
driver.execute_script("arguments[0].click();", category2)
sleep(1)

monthly_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[1]/div/ul/li[3]/a'
monthly = driver.find_element(By.XPATH, monthly_url)
driver.execute_script("arguments[0].click();", monthly)
sleep(1)

period_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[1]/span/label[3]'
period = driver.find_element(By.XPATH, period_url)
driver.execute_script("arguments[0].click();", period)
sleep(1)

device_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[3]/div/div/span[1]/label'
device = driver.find_element(By.XPATH, device_url)
driver.execute_script("arguments[0].click();", device)

gender_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[4]/div/div/span[1]/label'
gender = driver.find_element(By.XPATH, gender_url)
driver.execute_script("arguments[0].click();", gender)

age_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[5]/div/div/span[1]/label'
age = driver.find_element(By.XPATH, age_url)
driver.execute_script("arguments[0].click();", age)
sleep(1)

submit = driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div/div[1]/div/a')
driver.execute_script("arguments[0].click();", submit)
sleep(1)

male_btn_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[4]/div/div/span[2]/label'
male_btn = driver.find_element(By.XPATH, male_btn_url)

female_btn_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[4]/div/div/span[3]/label'
female_btn = driver.find_element(By.XPATH, female_btn_url)

age_btn_url = '//*[@id="content"]/div[2]/div/div[1]/div/div/div[5]/div/div/span/label'
age_btn = driver.find_elements(By.XPATH, age_btn_url)

# 데이터를 저장할 리스트 생성
all_rank = list()
male_list = [{'10대' : [], '20대':[], '30대':[], '40대':[], '50대':[], '60대 이상':[]}]
female_list = [{'10대' : [], '20대':[], '30대':[], '40대':[], '50대':[], '60대 이상':[]}]

# 랭킹 리스트 크롤링
def rank_crawling(save_list):
    rank_list = driver.find_elements(By.CSS_SELECTOR, '.rank_top1000_list li')
    for rank in rank_list:
        save_list.append(rank.text.split('\n')[1])
        
rank_crawling(all_rank)

driver.execute_script("arguments[0].click();", age)
driver.execute_script("arguments[0].click();", male_btn)

for idx, a_btn in enumerate(age_btn):
    if idx==0:
        pass
    else:
        driver.execute_script("arguments[0].click();", a_btn)
        driver.execute_script("arguments[0].click();", submit)
        sleep(5)
        rank_crawling(male_list[0][a_btn.text])
        driver.execute_script("arguments[0].click();", a_btn)
        
driver.execute_script("arguments[0].click();", male_btn)

driver.execute_script("arguments[0].click();", female_btn)

for idx, a_btn in enumerate(age_btn):
    if idx==0:
        pass
    else:
        driver.execute_script("arguments[0].click();", a_btn)
        driver.execute_script("arguments[0].click();", submit)
        sleep(5)
        rank_crawling(female_list[0][a_btn.text])
        driver.execute_script("arguments[0].click();", a_btn)
        
driver.execute_script("arguments[0].click();", female_btn)

driver.close()
```
